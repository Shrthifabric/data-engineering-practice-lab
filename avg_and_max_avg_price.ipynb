```python
from pyspark.sql.functions import col,avg,max
data = [ 
("2023-01-01","AAPL",150.00), 
("2023-01-02","AAPL",155.00), 
("2023-01-01","GOOG",2500.00), 
("2023-01-02","GOOG",2550.00), 
("2023-01-01","MSFT",300.00), 
("2023-01-02","MSFT",310.00) 
]
df=spark.createDataFrame(data,["date","stock","price"])
group_df=df.groupBy("date","stock").agg(avg(col("price")).alias("avg_price"))
max_avg_df=group_df.groupBy("stock").agg(max(col("avg_price")).alias("max_avg_price"))
max_avg_df.display()
```
