```python
dataDictionary = [ 
        ('James',{'hair':'black','eye':'brown'}), 
        ('Michael',{'hair':'brown','eye':None}), 
        ('Robert',{'hair':'red','eye':'black'}), 
        ('Washington',{'hair':'red','eye':'grey'}), 
        ('Jefferson',{'hair':'red','eye':''}) 
        ]
```

```python
from pyspark.sql.types import StructField, StructType, StringType, MapType
schema=StructType([StructField("name",StringType(),True),StructField("properties",MapType(StringType(),StringType()),True)])
df=spark.createDataFrame(dataDictionary,schema)
df=df.withColumn("hair",df.properties.getItem("hair")).withColumn("eye",df.properties.getItem("eye"))
df.display()
```
