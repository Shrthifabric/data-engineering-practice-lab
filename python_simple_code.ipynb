```python
def pyramid(n):
    for i in range(n):
        for j in range(i,n):
            print(" ",end="")
        for j in range(i+1):
            print("*",end="")
        for j in range(i):
            print("*",end="")
        print()
pyramid(5)

    
```

```python
def is_prime(num):
    for i in range(2,num):
        if num%i==0:
            return False
        else:
            return True
        
def generate_prime(n):
    num=2
    while n:
        if is_prime(num):
            yield num
            n-=1
        num+=1
x=int(input('enter a number of primenumbers to be generated'))
it=generate_prime(x)
for i in it:
    print(i,end=" ")


        
```

```python
def is_prime(n):
    flag=False
    if n<2:
        return flag
    for i in range(2,n):
        if n%i==0:
            return False
        else:
            flag=True
            return flag
        
n=int(input('enter a number'))
print(is_prime(n))
        
```

```python
def factorial(n):
    factor=1
    if n==0:
        #print('factorial of 0 is 1')
        return factor
    else:
        for i in range(1,n+1):
            factor=factor*i
        return factor
print(factorial(5))
```
